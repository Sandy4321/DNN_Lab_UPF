```python
import numpy as np
import cPickle as pkl
import os

from plotly import __version__
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot
init_notebook_mode()
```

```python
train_log_dir = '../complete/char_rnn/tiny_recipies/'
traces = []
for f in os.listdir(train_log_dir):
    model = f.split('_')[1]
    log = pkl.load(open(os.path.join(train_log_dir, f), 'rb'))
    train_cost = log['train_cost_history']
    traces.append(go.Scatter(
        x = np.arange(len(train_cost)),
        y = train_cost,
        name = model))
data = go.Data(traces)
iplot(data)
```
